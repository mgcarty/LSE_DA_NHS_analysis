```python
# Create a new DataFrame containing only the text.
tw_text = tw[['tweet_full_text']]

# View the DataFrame.
tw_text.head(5)
```

```python
# Create an empty list with the name "tags"
tags = []
```

```python
# Loop through the messages, and create a list of values containing the # symbol.
for y in [x.split(' ') for x in tw_text['tweet_full_text'].values]:
    for z in y:
        if '#' in z:
            # Change to lowercase.
            tags.append(z.lower())
            
# Create series 
tags_series = pd.Series(tags).value_counts()
```

```python
# Display the first 30 records.
tags_series.head(30)
```

```python
# Convert the series to a DataFrame in preparation for visualisation.
tags_data = pd.DataFrame(tags_series).reset_index()

tags_data.head(31)
```

```python
#Rename columns
tags_data.rename( columns={'index':'words', 0:'count'}, inplace=True)

tags_data
```

```python
# Fix the count datatype.
# View the result.

tags_data.dtypes
```

```python
# Create a Seaborn barplot indicating records with a count >10 records.

tags_grp = tags_data.groupby([tags_data.index,'words','count'])
tags_data_fil = tags_grp.filter(lambda x: x['count'] > 10.)

tags_data_fil
```

```python
# Create the plot.
# View the barplot.

tags_data_fil_excl = tags_data_fil.drop([0, 1])

sns.barplot(x=tags_data_fil_excl['count'], y=tags_data_fil_excl['words'])
plt.title('Hashtags with a count greater than 10', size=18)
plt.xlabel('Instances')
plt.ylabel('#Hashtags')
plt.savefig('5.png')
```
